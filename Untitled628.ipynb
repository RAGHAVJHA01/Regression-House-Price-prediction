```python
#set the data on parameter
#set the data o standerd scale to set the parameter
x=house.iloc[:,0:5]
y=house.iloc[:,-1]
```

```python
#set the parameter of the data iplement on standerd sc
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.fit(X_test,y_test)
```

```python
lr.score(X_test,y_test)
```

```python
#model evlotion
lr.intercept_
```

```python
lr.coef_
```

```python
predict=lr.predict(X_test)
predict
```

```python
plt.scatter(y_test,predict)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
#check the data set normally distrbutted or not
resudal=y_test-predict
resudal
sns.distplot(resudal,bins=50,color='b',hist_kws=dict(edgecolor='black',linewidth=1))
```

```python
#evalute the model
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predict))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
